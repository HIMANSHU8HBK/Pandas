```python
import pandas as pd

# Data for the Employees table
employees_data = {
    'emp_id': [1, 1, 1, 2, 2],
    'event_day': ['2020-11-28', '2020-11-28', '2020-12-03', '2020-11-28', '2020-12-09'],
    'in_time': [4, 55, 1, 3, 47],
    'out_time': [32, 200, 42, 33, 74]
}

# Create the DataFrame
employees = pd.DataFrame(employees_data)

# Display the DataFrame
employees
```

```python
employees = pd.read_excel("Data.xlsx", sheet_name = 'Day 21')
employees
```

```python
# Calculate time spent for each entry
employees['time_spent'] = employees['out_time'] - employees['in_time']
employees
```

```python
# Group by emp_id and event_day, then sum the time_spent
result_df = employees.groupby(['emp_id', 'event_day'], as_index=False)['time_spent'].sum()
result_df
```

```python
# Rename columns as required
result_df.rename(columns={'event_day': 'day', 'time_spent': 'total_time'}, inplace=True)
result_df
```
